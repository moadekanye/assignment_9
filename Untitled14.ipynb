```python
import seaborn as sns
import matplotlib.pyplot as plt

# Now load the planets dataset
planets = sns.load_dataset('planets')

# Example plot using the planets dataset
sns.relplot(data=planets, x='distance', y='orbital_period', hue='method', aspect=1.5)
plt.title("Orbital Period vs. Distance by Discovery Method")
plt.xlabel("Distance")
plt.ylabel("Orbital Period")
plt.legend(title="Discovery Method")
plt.show()

planets = sns.load_dataset('planets')
sns.relplot(data=planets, x='distance', y='orbital_period', hue='method', aspect=1.5)
plt.title("Orbital Period vs. Distance by Discovery Method")
plt.xlabel("Distance")
plt.ylabel("Orbital Period")
plt.legend(title="Discovery Method")
plt.show()

sns.relplot(data=planets, x='year', y='distance', hue='method', kind='scatter', aspect=1.5)
plt.title("Year of Discovery vs. Distance")
plt.xlabel("Year")
plt.ylabel("Distance")
plt.legend(title="Discovery Method")
plt.show()
plt.figure(figsize=(10, 6))
sns.histplot(data=planets, x='orbital_period', bins=30, kde=True)
plt.title("Distribution of Orbital Periods")
plt.xlabel("Orbital Period")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=planets, x='method', y='distance')
plt.title("Distribution of Distances by Discovery Method")
plt.xlabel("Discovery Method")
plt.ylabel("Distance")
plt.xticks(rotation=45)
plt.show()
plt.figure(figsize=(10, 6))
sns.countplot(data=planets, x='method')
plt.title("Count of Planets by Discovery Method")
plt.xlabel("Discovery Method")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(data=planets, x='year', y='distance', estimator='mean', ci=None)
plt.title("Average Distance by Year of Discovery")
plt.xlabel("Year")
plt.ylabel("Average Distance")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Exercise_Data.csv')

# Pivot the data to create a matrix for the heatmap
data_pivot = data.pivot(index='diet', columns='exercise_type', values='pulse')

# Create the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data_pivot, cmap='coolwarm', annot=True)
plt.title("Heatmap of Pulse Data by Diet and Exercise Type")
plt.xlabel("Type of Exercise")
plt.ylabel("Diet")
plt.show()
```
